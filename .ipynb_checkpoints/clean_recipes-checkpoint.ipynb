```python
import pandas as pd
import re

# Load the file
df = pd.read_csv("cleaned_recipes.csv", encoding='ISO-8859-1')
```

```python
print(df)
```

```python
print(df.columns)
```

```python
def clean_keywords(text):
    if pd.isna(text):
        return ''
    text = text.lower()
    text = re.sub(r'\b\d+(\.\d+)?\b', '', text)  # remove plain numbers
    text = re.sub(r'[½¼¾⅓⅔⅛⅜⅝⅞]', '', text)        # remove fractions
    text = re.sub(r'\b(cups?|tablespoons?|teaspoons?|oz|ounces?|grams?|g|lbs?|pounds?)\b', '', text)
    text = re.sub(r'[^a-zA-Z,\s]', '', text)     # remove anything not a-z or space or comma
    text = re.sub(r'\s+', ' ', text)             # collapse extra spaces
    return text.strip()

df['Keywords'] = df['Keywords'].apply(clean_keywords)

df.to_csv("keywords_recipes.csv", index=False)
```
